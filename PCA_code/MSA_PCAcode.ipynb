### PCA for Data Visualization

Analyze a microarray gene expression dataset using the principle component analysis program:

### Importing the modules

```python
#importing all the required modules for PCA
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

### Loading the data

The dataset contains 4298 rows/genes and 380 columns/growth conditions.

```python
#Reading the text file into CSV dataframe using pandas
#file output is not shown as it is a large dataset.
ecoli_dataset = pd.read_csv('C:/Users/mayur/Documents/UNCC/MSA/avg_E_coli_exp380_clean.dat.txt', header = None, sep = "\t")
```

### Standarizing the data

PCA is effected by scale so need to scale the features in the data before applying PCA. Use StandardScaler to help standardize the datasetâ€™s features onto unit scale (mean = 0 and variance = 1) which is a requirement.

```python
#standardscaler stored into a variable
scaler = StandardScaler()

#Standardizing the features using fit transform
ecoliData_scaled = scaler.fit_transform(ecoli_dataset)
ecoliData_scaled
```

### PCA projection to 2D

```python
#assigning the PCA to 2 dimensional
pca = PCA(n_components=2)

#applying the pca to scaled dataset and transformed into two dimensional
pca.fit(ecoliData_scaled)
principalComponents = pca.fit_transform(ecoliData_scaled)

#shape helps to view the data dimensions reduction from original dimension
ecoliData_scaled.shape, principalComponents.shape
```

###  Visualization in 2D dimension

```python
#plotting the scatter plot to view data in 2D.
plt.figure(figsize=(12,8)) #setting the figure size
plt.scatter(principalComponents[:,0],principalComponents[:,1]) #plotting the data of both components
# labelling the axes and title
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("2 component of PCA", fontsize = 20) #fontsize of title
```

### Explained Variance

The explained variance tells you how much information (variance) can be attributed to each of the principal components. 

```python
#variance ratio of first two components.
pca.explained_variance_ratio_
```

```python
#sum of the variance of 2 components.
pca.explained_variance_ratio_.sum()
```

By using the attribute explained_variance_ratio_, we can see that the first principal component contains 80.2% of the variance and the second principal component contains 4.82% of the variance. Together, the two components contain 85% of the information.
